```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

# Acesso componentes do SISTEMA OPERACIONAL
import os 
from pathlib import Path
```

```python
# DATA BASE
Local_Diretorio = 'C:/Users/nilva/OneDrive/Área de Trabalho/Dashboard_multas_detran/ajustados_2022'
Diretorio_Path = Path(Local_Diretorio)

type(Diretorio_Path)
```

# Lista os arquivos

```python
Arquivos = Diretorio_Path.iterdir()

# Função para calcular o tamanho do arquivo
def Bytes_Mega(bytes):
    Megabytes = bytes / ( 1024 * 1024)
    return Megabytes

# total MBs
Total_MB = 0

# Loop
for Loop in Arquivos:
 
    
# Busca o tamanho do arquivo
    Tamanho = os.path.getsize ( Loop )
    
# Convertendo em Megas
    Conversao = Bytes_Mega( Tamanho )
    
# Extraindo o nome do arquivo
    
    Nome_Arquivo = str( Loop ).split('\\')[-1]
    
    print(f' {Nome_Arquivo}  : {round (Conversao, 2) } MB')
    print ('-' * 30 )
    
    Total_MB = Total_MB + Tamanho 
    
    
print(f'Temos ao todo { round( Bytes_Mega(Total_MB), 2) } MBs de dados' )
    
    
```

```python
pd.read_csv( Loop, error_bad_lines=False, sep=';', encoding='latin-1' )
```

```python
# Listar os arquivos
Arquivos = Diretorio_Path.iterdir()

# Loop para consolidar os Arquivos
for Contador, Arq in enumerate(Arquivos):
    
    print(Contador, Arq )
    
# Se for a 1° interação, cria a base consolidada
    if Contador == 0:
    # Ler o arquivo do Loop
        Base_Dados = pd.read_csv( Arq, error_bad_lines=False, sep=';', encoding='latin-1' )
    #Salvar na variavel
        Base_Consolidada = Base_Dados
    
    else:
        # ler o arquivo no loop
        Base_Dados = pd.read_csv( Arq, error_bad_lines=False, sep=';', encoding='latin-1' )
        
        # Empilhar os dados
        Base_Consolidada = pd.concat( [Base_Consolidada, Base_Dados] )
        
# Olhar para a dimensão
Base_Consolidada.shape
```

# Exportando Dados
